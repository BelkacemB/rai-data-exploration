## Correlations

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = df.loc[:, 'duration_ms_x':'youtube_views']
df = df.infer_objects()
corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype=np.bool))
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
ax = sns.heatmap(corr, mask=mask, cmap=cmap, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
ax.set_title("Chaabi songs - Audio features and popularity - Correlation matrix")
```
